```python
#tutorial elliptic c over relaxation technique (relax lah sangat)
import numpy as np
from sympy import symbols, Eq, linear_eq_to_matrix
from scipy.linalg import solve

n   = 3
UII  = []
u   = []
v   = []
X   = []

for j in range(0,n+1):
    for i in range(0,n+1):
        u.append(str('U')+str(j)+str(i))
        if   (j >0 and j < (n)) and (i>0 and i<(n)):
            v.append(str('U')+str(j)+str(i))      
        else:
            None
            
u = symbols(u)
U = np.reshape(u, (n+1,n+1)) 

for i in range (1,n):
    U[i,0] = 0 # y = 0 boundary (bawah)
    U[i,n] = 75 # y = n boundary (atas)
    U[0,i] = 50 # x = 0 boundary (kiri)
    U[n,i] = 25 # x = n boundary (kanan)

for i in range(1,n):
    for j in range(1,n):
        UI = (U[i+1,j]+U[i,j+1]+U[i,j-1]+U[i-1,j])/4
        UII.append(UI)
        

v = symbols(v) 

print(UII)
print()

U1 = lambda U11,U12,U21,U22: U12/4 + U21/4 + 25/2
U2 = lambda U11,U12,U21,U22: U11/4 + U22/4 + 125/4
U3 = lambda U11,U12,U21,U22: U11/4 + U22/4 + 25/4
U4 = lambda U11,U12,U21,U22: U12/4 + U21/4 + 25

U110 = 0
U120 = 0
U210 = 0
U220 = 0 
count = 1

# Reading tolerable error
e = 0.0001

# Reading relaxation factor
w = 1.5

condition = True

while condition:
    U111 = (1-w) * U110 + w * U1(U110,U120,U210,U220)
    U121 = (1-w) * U120 + w * U2(U111,U120,U210,U220)
    U211 = (1-w) * U210 + w * U3(U111,U121,U210,U220)
    U221 = (1-w) * U220 + w * U4(U111,U121,U211,U220)
    print('%d\t%0.4f\t%0.4f\t%0.4f\t%0.4f\n' %(count, U111,U121,U211,U221))
    e1 = abs(U110-U111);
    e2 = abs(U120-U121);
    e3 = abs(U210-U211);
    e4 = abs(U220-U221);
    
    count += 1
    U110 = U111
    U120 = U121
    U210 = U211
    U220 = U221
    
    condition = e1>e and e2>e and e3>e and e4>e

print((U111,U121,U211,U221))
```
